```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression 
```

```python
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
#labels: 0 = Fail, 1 = Pass
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])
```

```python
# Print to verify
print("X:\n", X)
print("y:\n", y)
```

```python
model = LogisticRegression()
model.fit(X,y)
```

```python
X_test = np.linspace(0, 9, 100).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='black', label="Data Points (Fail=0, Pass=1)")
plt.plot(X_test, y_prob, color='blue', label="LogisticRegresion Curve")
plt.axhline(0.5, color='red', linestyle='--', label="Threshold = 0.5")
plt.title("Logistic Regression: Study Hours vs. Pass/fail")
plt.xlabel("Study Hours")
plt.ylabel("probablity of passing")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# Predict using the fitted polynomial model
X_test = np.linspace(0, 9, 100).reshape(-1, 1)
X_test_poly = poly_features.transform(X_test)
y_pred = model.predict(X_test_poly)
```

```python
# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='black', label="Data Points (Fail=0, Pass=1)")
plt.plot(X_test, y_pred, color='pink', label="Polynomial Regression (Degree 2)")
plt.title("Polynomial Regression: Study Hours vs. Pass/Fail")
plt.xlabel("Study Hours")
plt.ylabel("Predicted Value")
plt.legend()
plt.grid(True)
plt.show()
```
